```python
import re
```

```python
def read_register(file_path):
    register = []
    data = r'(?P<ip>\d+\.\d+\.\d+\.\d+).*?\[(?P<date>.*?)\] "(?P<method>\w+) (?P<route>.*?) HTTP/.*?" (?P<status>\d+)'
    
    with open(file_path, 'r') as log_file:
        for line in log_file:
            match = re.search(data, line)

            if match:
                register.append(match.group("ip"))
                register.append(match.group("date"))
                register.append(match.group("method"))
                register.append(match.group("route"))
                register.append(match.group("status"))
                
                
    return register

    
```

```python
import re

def browser_file(file_path):
    create_dict = []
    info = r'(?P<user_agent>Mozilla/[^"]+)' 

    with open(file_path, 'r') as log_file:
        for line in log_file:
            match = re.search(info, line)

            if match:
                create_dict.append({
                    "Browser": match.group('user_agent') 
                })
    return create_dict
```

```python
file_path = "primer_registro_log.txt"
records = read_register(file_path)
browsers = browser_file(file_path)
```

```python
for register in records[:5]:
    print(register)
```

```python
for browser in browsers:
    print(browser)
```
